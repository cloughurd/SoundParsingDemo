```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt

ambience, sr_a = sf.read('ambience.wav')
plt.plot(ambience)
plt.show()

ggb, sr_g = sf.read('gingerbread.wav')
plt.plot(ggb)
plt.show()
```

```python
max_amb = np.percentile(np.abs(ambience), 99.9)
max_amb
```

```python
plt.plot(ggb)
plt.show()

plt.plot(ggb)
plt.xlim(9000, 10000)
plt.show()
```

```python
len(ggb)
```

```python
pooled = ggb.copy()
pool_size = 20
for i in range(pool_size, len(ggb) - pool_size):
    pooled[i] = np.max(np.abs(ggb[i - pool_size: i + pool_size]))
    
plt.plot(pooled)
plt.show()
```

```python
pooled = np.where(pooled > max_amb, 1, 0)
word_duration = np.sum(pooled) / sr_g * 1000
wsd = word_duration / 3
wsd
```
