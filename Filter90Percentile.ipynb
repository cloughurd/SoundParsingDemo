```python
import matplotlib.pyplot as plt
import soundfile as sf

data, fs = sf.read('gingerbread2019_09_30_10_01_39.wav')
data
```

```python
plt.plot(data)
plt.show()
```

```python
in_seconds = [x/fs * 1000 for x in range(len(data))]
plt.plot(in_seconds, data)
plt.show()
```

```python
import numpy as np

data_c = data.copy()
np.mean(data_c)
```

```python
np.percentile(np.abs(data), 90)
```

```python
data_c = np.where(np.abs(data) < 0.026, 0, data)
plt.plot(data_c)
plt.show()
```

```python
sf.write('quieted.wav', data_c, fs)
```

```python
binary_data = np.where(np.abs(data) < 0.026, 0, 1)
word_duration = np.sum(binary_data) / fs * 1000
wsd = word_duration / 3
wsd
```

```python
plt.plot(binary_data)
plt.plot(data, color='r')
plt.ylim((0,0.4))
plt.show()
```
